```python
def precision(row):
    return row['tp']/(row['fp']+row['tp'])

def recall(row):
    return row['tp']/(row['tp']+row['fn'])

def test(row):
    print(row['tp'])
    return row
```

```python
for colour in df.columns.levels[0]:    
    df[(colour, 'precision')] = df[colour][['tp', 'fp']].apply(precision, axis=1)
    df[(colour, 'recall')] = df[colour][['tp', 'fn']].apply(recall, axis=1)
```
