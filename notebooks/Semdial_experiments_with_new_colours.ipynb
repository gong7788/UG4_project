```python
yellow_on_blue = rules.Rule.generate_red_on_blue_options(['yellow'], ['blue'])
purple_on_yellow = rules.Rule.generate_red_on_blue_options(['purple'], ['yellow'])
red_on_yellow = rules.Rule.generate_red_on_blue_options(['red'], ['yellow'])
blue_cm = pgm_agent.add_cm('blue')
yellow_cm = pgm_agent.add_cm('yellow')
red_cm = pgm_agent.add_cm('red')
purple_cm = pgm_agent.add_cm('purple')
```

```python
teacher_ = teacher.TeacherAgent()
pgm_agent = PGMAgent.PGMCorrectingAgent(w, teacher=teacher_)
pgm_agent.pgm_model.known_rules = set(red_on_yellow).union(set(purple_on_yellow))
pgm_agent.pgm_model.colours['red'] = red_cm
pgm_agent.pgm_model.colours['purple'] = purple_cm
do_model_update(pgm_agent, ['b6', 'b7', 'b0'], yellow_on_blue, yellow_cm, blue_cm,
               0, table_correction=True, no_correction=False)
do_model_update(pgm_agent, ['b5', 'b7', 'b0'], yellow_on_blue, yellow_cm, blue_cm,
               1, table_correction=True, no_correction=False)
do_model_update(pgm_agent, ['b4', 'b6', 'b0'], yellow_on_blue, yellow_cm, blue_cm,
               2, table_correction=True, no_correction=False)
do_model_update(pgm_agent, ['b2', 'b6', 'b0'], blue_on_yellow, blue_cm, yellow_cm,
               3, table_correction=True, no_correction=False)
do_model_update(pgm_agent, ['b9', 'b7', 'b0'], blue_on_yellow, blue_cm, yellow_cm,
               4, table_correction=True, no_correction=False)
do_model_update(pgm_agent, ['b3', 'b8'], blue_on_yellow, blue_cm, yellow_cm,
               5, table_correction=True, no_correction=True)

print(len(pgm_agent.pgm_model.search_inference.beam))
```

```python
def do_model_update(pgm_agent, args, rules, blue_cm, yellow_cm, time, 
                    table_correction=False, no_correction=False):
    corr = f'corr_{time}'
    data = pgm_agent.get_colour_data(args)
    pgm_agent.time = time
    if no_correction:
        pgm_agent.pgm_model.add_no_correction(args, time)
        data[corr] = 0
    else:
        pgm_agent.pgm_model.extend_model(rules, blue_cm, yellow_cm, args, time, 
                                 table_correction=table_correction)
        data[corr] = 1
    pgm_agent.pgm_model.observe(data)
    
```
